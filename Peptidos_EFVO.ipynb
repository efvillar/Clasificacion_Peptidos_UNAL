## Preprocesamiento

```python
data.shape
```

```python
# Separaci√≥n de los datos en train y test
# random_state para reproducibilidad

X_train, X_test, y_train, y_test = train_test_split(data, data.Respuesta,
                                                    test_size=0.2,
                                                    random_state=0)
X_train.shape, X_test.shape
```

```python
#varaibles numericas a estandarizar
train_vars = [var for var in X_train.columns if var not in ['sequence']]
len(train_vars)
```

```python
#ajuste del escalador
scaler = StandardScaler()  #se crea una instancia del escalador
scaler.fit(X_train[train_vars]) #  se ajusta el escalador a los datos de entrenamiento

# Se transforman los datos de train y test con el escalador ajustado a los datos de entrenamiento
X_train = pd.concat([X_train[['sequence']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(X_train[num_vars]), columns=train_vars)],
                    axis=1)

X_test = pd.concat([X_test[['sequence']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(X_test[num_vars]), columns=train_vars)],
                    axis=1)
```

```python
# Se guardan los datos estandarizados de test y train

X_train.to_pickle('./Datos/xtrain.pkl')
X_test.to_pickle('./Datos/xtest.pkl')
y_train.to_pickle('./Datos/ytrain.pkl')
y_test.to_pickle('./Datos/ytest.pkl')
```

## Selecci√≥n y Reducci√≥n de Variables

```python
#se cargan los datos estandarizados
X_train = pd.read_pickle('./Datos/xtrain.pkl')
X_test = pd.read_pickle('./Datos/xtest.pkl')
y_train = pd.read_pickle('./Datos/ytrain.pkl')
y_test = pd.read_pickle('./Datos/ytest.pkl')
```

Se usar√° una regresi√≥n lasso,, con una semilla = 0, y la funci√≥n selectFromModel de sklearn para seleccionar las varaibles m√°s predictivas.  Entre mayor alpha m√°s reducci√≥n de variables

```python
#lr = LogisticRegression(solver='liblinear')
```

```python
X_train.head(2)
```

```python
X_train.drop(['sequence', 'Respuesta'], axis=1, inplace=True)
```

```python
sel_ = SelectFromModel(Lasso(alpha=0.05, random_state=0)) # remember to set the seed, the random state in this function
sel_.fit(X_train, y_train)
```

```python
sel_.get_support()  #para visualizar las varaibles seleccionas
```

```python
sum(sel_.get_support())  #se seleccionaron 15 variables significativas seg√∫n el modelo RL Lasso
```

```python
selected_feat = X_train.columns[(sel_.get_support())]
```

```python
selected_feat  #caracteristicas seleccionadas
```

```python
# algunas estadisticas
print('total variables: {}'.format((X_train.shape[1])))
print('varaibles seleccionadas: {}'.format(len(selected_feat)))
print('Variables excluidas por RL Lasso: {}'.format(
    np.sum(sel_.estimator_.coef_ == 0)))
```

```python
selected_feats = X_train.columns[(sel_.estimator_.coef_ != 0).ravel().tolist()]
selected_feats  #otra forma de ver las caracteriticas seleccionadas
```

```python
# nse guarda la lista de las caracteristicas seleccionadas
pd.Series(selected_feats).to_csv('./Datos/selected_features.csv', index=False)
```
